Freeze UCI - UCI

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega o arquivo CSV em um DataFrame
df = pd.read_csv('/workspaces/betania.silva/ssl_tools/contrastive-predictive-coding-for-har/freeze_uci_uci.csv')

# Filtra apenas as entradas de treinamento
train_df = df[df['Phase'] == 'train']

# Filtra apenas as entradas de teste
test_df = df[df['Phase'] == 'test']

val_df = df[df['Phase'] == 'val']

# Plota o gráfico de evolução da loss durante o treinamento e teste
plt.plot(train_df['Epoch'], train_df['Loss'], label='Train Loss')
plt.plot(test_df['Epoch'], test_df['Loss'], label='Test Loss')
plt.plot(val_df['Epoch'], val_df['Loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Evolução da Loss durante o Treinamento e Teste')
plt.legend()
plt.show()
```

Pretrain UCI 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega o arquivo CSV em um DataFrame
df = pd.read_csv('/workspaces/betania.silva/ssl_tools/contrastive-predictive-coding-for-har/loss_uci_pretrain.csv')

# Filtra apenas as entradas de validação
val_df = df[df['Phase'] == 'val']
train_df = df[df['Phase'] == 'train']

# Plota o gráfico de evolução da loss durante a validação
plt.plot(val_df['Epoch'], val_df['Loss'], label='Validation Loss')
plt.plot(train_df['Epoch'], train_df['Loss'], label='Train Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Evolução Loss de validação durante o treinamento')
plt.legend()
plt.show()
```
